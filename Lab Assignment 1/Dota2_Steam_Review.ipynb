```python
# Name: JEEVANRAJ A/L THAYANITHI
# ID: SW01083389
# Name: MOHAMAD SHAFIQ BIN MOHAMAD AZMI
# ID: IS01083305
# CISB5123 Text Analytics - Lab Assignment 1

import requests
import pandas as pd
from datetime import datetime

# Configuration
app_id = 570  # Dota 2 App ID
url = f"https://store.steampowered.com/appreviews/{app_id}"
all_reviews = []
cursor = "*" 

# The assignment requires a limit of 5 pages 
pages_to_scrape = 5
reviews_per_page = 20 

print(f"Starting scrape for {pages_to_scrape} pages...")

for page in range(pages_to_scrape):
    # Parameters for the Steam API request
    params = {
        "json": 1,
        "num_per_page": reviews_per_page,
        "cursor": cursor,
        "filter": "recent",
        "language": "english"
    }
    
    # Headers make the request look like it's coming from a browser (Best Practice)
    headers = {'User-Agent': 'Mozilla/5.0'}
    
    response = requests.get(url, params=params, headers=headers)
    data = response.json()
    
    if not data.get("reviews"):
        break
    
    # Extracting required fields: Name, Date, and Content [cite: 10, 11, 12, 13]
    for review in data["reviews"]:
        all_reviews.append({
            "Reviewer Name": review["author"]["steamid"], # Reviewer ID/Name [cite: 11]
            "Review Date": datetime.fromtimestamp(review["timestamp_created"]).strftime("%Y-%m-%d"), # Date [cite: 12]
            "Review Content": review["review"] # Content [cite: 13]
        })
    
    # Update cursor to get the next page of results 
    cursor = data.get("cursor")
    print(f"Page {page + 1} scraped successfully.")

# Save to CSV using utf-8-sig to handle special characters 
df = pd.DataFrame(all_reviews)
df.to_csv("steam_reviews.csv", index=False, encoding='utf-8-sig')

print("-" * 30)
print(f"Scraping completed. Total reviews saved: {len(df)}")
```
